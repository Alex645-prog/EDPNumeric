```
import numpy as np
import matplotlib.pyplot as plt
```

```
# inputs
a = 0
b = 2
c = 0
d = 1
m = 5
n = 6
tol = 10**(-10)
max_iter = 100
tol
```

```
# paso 1
h = float((b-a))/n
k = float((d-c))/m
print(h,k)
```

```
# paso 2  y 3
x = np.zeros(n+1)
y = np.zeros(m+1)

# i = 0:n , sumamos 1 al final 
#paso 2
for i in range(n+1): x[i] = a + i*h
#paso 3
for j in range(m+1): y[j] = c + j*k
print(x)
print(y)
```

```
## numeracion de una matriz

w = np.zeros((n+1,m+1))
for i in range(n+1):
  for j in range(m+1):
    w[i,j] = x[i]*np.exp(y[j])#i+float(j)/10

np.around(w,decimals=5)
```

```
# valores iniciales
w = np.zeros((n+1,m+1)) # para que cada columna represente un punto en x

# paso 4
# DEFINIR EL ORDEN : en este caso de abajo a arriba
for i in range(n+1):
  w[i,0] = x[i]  # u(x,0)
  w[i,m] = np.e*x[i]  # u(x,1)
for j in range(m+1):
  w[0,j] = 0   # u(0,y)
  w[n,j] = 2*np.exp(y[j])   # u(2,y)

np.around(w,decimals=5)
```

```
# paso 5
r = round(h**2/k**2,5)
u = round(2*(1+r),5)
print(r,u)
```

```
# definir el f(x,y)
def f(x,y):
  return x*np.exp(y)

# definir el g(x,y)
def g(x,y):
  return x*np.exp(y)
```

[Gauss Seidel](https://www.maa.org/press/periodicals/loci/joma/iterative-methods-for-solving-iaxi-ibi-gauss-seidel-method)

```
z = (-h**2*f(x[1],y[m-1])+g(a,y[m-1])+r*g(x[1],d)+r*w[1,m-2]+w[2,m-1])/u
norm = abs(z-w[1,m-1])
w[1,m-1] = z
w
```

```
for i in range(2,n-1):
    z = (-h**2*f(x[i],y[m-1])+r*g(x[i],d)+w[i-1,m-1]+w[i+1,m-1]+r*w[i,m-2])/u
    if abs(w[i,m-1] - z) > norm:
      norm = abs(w[i,m-1] - z)
    w[i,m-1] = z
w
```

```
z = (-h**2*f(x[n-1],y[m-1])+g(b,y[m-1])+r*g(x[n-1],d)+w[n-2,m-1]+r*w[n-1,m-2])/u
if abs(w[n-1,m-1] - z) > norm:
    norm = abs(w[n-1,m-1] - z)
w[n-1,m-1] = z
w
```

```
#paso 10
# para j=2:-1:m-2 , se debe restar 1 al final para el range
for j in range(m-2,1,-1):
  #paso 11
  z = (-h**2*f(x[1],y[j])+g(a,y[j])+r*w[1,j+1]+r*w[1,j-1]+w[2,j])/u
  if abs(w[1,j]-z) > norm:
    norm = abs(w[1,j]-z)
  w[1,j] = z
  #paso 12
  for i in range(2,n-1):
    z = (-h**2*f(x[i],y[j])+w[i-1,j]+r*w[i,j+1]+w[i+1,j]+r*w[i,j-1])/u
    if abs(w[i,j]-z) > norm:
      norm = abs(w[i,j]-z)
    w[i,j] = z
  #paso 13
  z = (-h**2*f(x[n-1],y[j])+g(b,y[j]) +w[n-2,j]+r*w[n-1,j+1]+r*w[n-1,j-1])/u
  if abs(w[n-1,j]-z) > norm:
    norm = abs(w[n-1,j]-z)
  w[n-1,j] = z
w
```

```
# Paso 14
z =(-h**2*f(x[1],y[1])+ g(a,y[1])+r*g(x[1],c)+r*w[1,2]+w[2,1])/u
if abs(w[1,1]-z) > norm:
  norm = abs(w[1,1]-z)
w[1,1] = z
w
```

```
# Paso 15
for i in range(2,n-1):
  z = (-h**2*f(x[i],y[1])+r*g(x[i],c)+w[i-1,1]+r*w[i,2]+w[i+1,1])/u
  if abs(w[i,1]-z) > norm:
    norm = abs(w[i,1]-z)
  w[i,1] = z
w
```

```
z = (-h**2*f(x[n-1],y[1])+g(b,y[1])+r*g(x[n-1],c)+w[n-2,1]+r*w[n-1,2])/u
if abs(w[n-1,1]-z) > norm:
  norm = abs(w[n-1,1]-z)
w[n-1,1] = z
w
```

```
#paso 6
l = 1
while l <= max_iter :
  # paso 7 gauss seidel
  z = (-h**2*f(x[1],y[m-1])+g(a,y[m-1])+r*g(x[1],d)+r*w[1,m-2]+w[2,m-1])/u
  norm = abs(z-w[1,m-1])
  w[1,m-1] = z
  #paso 8
  for i in range(2,n-1):
    z = (-h**2*f(x[i],y[m-1])+r*g(x[i],d)+w[i-1,m-1]+w[i+1,m-1]+r*w[i,m-2])/u
    if abs(w[i,m-1] - z) > norm:
      norm = abs(w[i,m-1] - z)
    w[i,m-1] = z
  #paso 9 
  z = (-h**2*f(x[n-1],y[m-1])+g(b,y[m-1])+r*g(x[n-1],d)+w[n-2,m-1]+r*w[n-1,m-2])/u
  if abs(w[n-1,m-1] - z) > norm:
      norm = abs(w[n-1,m-1] - z)
  w[n-1,m-1] = z
  
  #paso 10
  # para j=2:-1:m-2 , se debe restar 1 al final para el range
  for j in range(m-2,1,-1):
    #paso 11
    z = (-h**2*f(x[1],y[j])+g(a,y[j])+r*w[1,j+1]+r*w[1,j-1]+w[2,j])/u
    if abs(w[1,j]-z) > norm:
      norm = abs(w[1,j]-z)
    w[1,j] = z
    #paso 12
    for i in range(2,n-1):
      z = (-h**2*f(x[i],y[j])+w[i-1,j]+r*w[i,j+1]+w[i+1,j]+r*w[i,j-1])/u
      if abs(w[i,j]-z) > norm:
        norm = abs(w[i,j]-z)
      w[i,j] = z
    #paso 13
    z = (-h**2*f(x[n-1],y[j])+g(b,y[j]) +w[n-2,j]+r*w[n-1,j+1]+r*w[n-1,j-1])/u
    if abs(w[n-1,j]-z) > norm:
      norm = abs(w[n-1,j]-z)
    w[n-1,j] = z
  # Paso 14
  z =(-h**2*f(x[1],y[1])+ g(a,y[1])+r*g(x[1],c)+r*w[1,2]+w[2,1])/u
  if abs(w[1,1]-z) > norm:
    norm = abs(w[1,1]-z)
  w[1,1] = z
  # Paso 15
  for i in range(2,n-1):
    z = (-h**2*f(x[i],y[1])+r*g(x[i],c)+w[i-1,1]+r*w[i,2]+w[i+1,1])/u
    if abs(w[i,1]-z) > norm:
      norm = abs(w[i,1]-z)
    w[i,1] = z

  # Paso 16
  z = (-h**2*f(x[n-1],y[1])+g(b,y[1])+r*g(x[n-1],c)+w[n-2,1]+r*w[n-1,2])/u
  if abs(w[n-1,1]-z) > norm:
    norm = abs(w[n-1,1]-z)
  w[n-1,1] = z

  if norm <= tol:
    print('termino en iteracion: ',l)
    break
  else:
    l = l + 1
```

### Metodo matricial usando librerias

```
def tridiag(a,b,c,N):
    A = np.zeros([N,N])
    np.fill_diagonal(A[1:,:-1],a)
    np.fill_diagonal(A[:-1,1:],c)
    np.fill_diagonal(A,b)
    return A
```

```
A = tridiag(-r,u,-r,4)
A
```
