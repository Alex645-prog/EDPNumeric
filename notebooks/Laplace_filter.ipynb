```python
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
```

```python
img = plt.imread('windows.jpg')
```

```python
plt.imshow(img)
plt.show()
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.114 , 0.587 , 0.299])
```

```python
gray = rgb2gray(img)
```

```python
plt.imshow(gray, cmap='gray')
```

```python
img_laplace = ndimage.filters.laplace(gray)
```

```python
plt.imshow(img_laplace, cmap='gray')
```

```python
img.shape
```

```python
np.max(img_laplace)
```

```python
def binary_image(gray , umbral):
    m,n = gray.shape
    bin_img = gray.copy()
    for i in range(m):
        for j in range(n):
            if gray[i][j] < umbral:
                bin_img[i][j] = 0
            else:
                bin_img[i][j] = 1
                
    return bin_img
```

```python
bin_image = binary_image(img_laplace, 50)
```

```python
plt.imshow(bin_image,cmap='gray')
```

```python
fr = ndimage.fourier.fourier_gaussian(gray,0.8)
```

```python
plt.imshow(fr, cmap='gray')
```
